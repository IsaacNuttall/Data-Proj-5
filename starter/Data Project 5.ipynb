```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('supermarket.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# import seaborn
import seaborn as sns

# set standard plot type as seaborn plot
sns.set()

# import matplotlib
import matplotlib.pyplot as plt

# write command for plots to appear in jupyter cells automatically
%matplotlib inline
```

```python
plt.figure(figsize=(12,4))

plt.title("Sales Distribution Among Stores", fontsize=16)

sns.boxplot(x=df['store_sales'])

plt.xlabel("Store Sales", fontsize=14);
```

```python
plt.figure(figsize=(12,4))

plt.title("Customer Distribution Among Stores", fontsize=16)

sns.boxplot(x=df['daily_customer_count'])

plt.xlabel("Store Sales", fontsize=14);
```

```python
plt.figure(figsize=(12,4))

plt.title("Inventory Distribution Among Stores", fontsize=16)

sns.boxplot(x=df['items_available'])

plt.xlabel("Store Inventory", fontsize=14);
```

```python
sns.lineplot(x="items_available", y="store_sales",
             hue="items_available", style="store_sales",
             data=df)
```
